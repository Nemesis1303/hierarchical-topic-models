```python
from src.evaluateMatching.alignment import Alignment

al = Alignment()

new_pairs, new_pairs2, new_pairs3, wmd = al.do_one_to_one_matching()
```

```python
import numpy as np
import plotly.express as px

x = ["Topic " + str(i) for i in range(wmd.shape[1])]
y = ["Topic " + str(i) for i in range(wmd.shape[0])]
z = wmd
z_text = np.around(z, decimals=2) # Only show rounded value (full value on hover)

fig = px.imshow(z, x=x, y=y, color_continuous_scale='Blues', aspect="auto")
fig.update_traces(text=z_text, texttemplate="%{text}")
fig.update_xaxes(side="top", tickangle=-45)  # Set the desired rotation angle
fig.show()
```
